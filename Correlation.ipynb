1.	Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

Covariance is a statistical measure that quantifies the relationship between two variables, like correlation. However, unlike Correlation, Covariance is influenced by the scale of the variables. Variables with larger magnitudes will have larger covariances, even if the relationship between them is the same. Correlation is not affected by the scale of the variable, as it is scale independence.

Covariance being unscaled, lacks a straightforward interpretation. Although, A positive covariance suggests that when one variable is above its mean, the other tends to be above its mean as well. A negative covariance suggests an inverse relationship whereas Correlation provides a clear interpretation of the relationship. A correlation coefficient of (+1) indicates a perfect positive relationship, (-1) indicates a perfect relationship, and 0 suggests no linear relationship.

2.	What does a positive, negative, and zero covariance indicate about the relationship between two variables?

A positive covariance means two variables move in the same direction (both increase/decrease); a negative covariance means they move in opposite directions (one increases, the other decreases); and a zero covariance indicates no linear relationship, suggesting they change independently.

3.	Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

Limitations:

It has units which complicates comparison between different variable pairs. Covariance is scale dependant. Its value changes with the units (e.g., meters vs. centimetres) or scale of the variables, making interpretation difficult. It has unstandardized magnitude.

Why is Correlation preferred?

It is standardized and unitless. It has a clear strength indication. It identifies direction and strength. Its comparability allows meaningful comparison of relationships between different pairs of variables or datasets.

4.	 Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?
Pearson's correlation measures linear relationships between continuous, normally distributed variables, sensitive to outliers, while Spearman's measures monotonic (consistent direction, not necessarily straight-line) relationships using ranks, robust to outliers, and suitable for ordinal or non-normal data.
Use Spearman's when data is ordinal, non-normally distributed, has outliers, or the relationship is monotonic but not perfectly linear, whereas Pearson's is for strong linear ties in clean, continuous data.
5.	 If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?

A correlation coefficient of 0.85 indicates a strong positive linear relationship as variable X increases, variable Y also tends to increase significantly, and vice versa, with the variables moving closely together. You cannot infer causation, because correlation only measures the strength and direction of a relationship, not that one variable causes the other; other factors (confounders) might influence both, or the relationship could be coincidental.
